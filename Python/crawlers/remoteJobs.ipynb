```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://remote.co/remote-jobs/developer/"
r = requests.get(url)
soup = bs(r.content, "lxml")
```

```python
remoteJobs = soup.find_all('div', class_='card-body p-0')
for job in remoteJobs:
    #print(job)
    titles_company = job.find_all('div', class_="card-body px-3 py-0 pl-md-0")
    for i in titles_company:
        #print(i)
        title = i.find('span', class_='font-weight-bold larger')
        company = i.find('p', class_='m-0 text-secondary')
        if title:
            print(title.text)
        if company:
            t = company.text.strip()
            print(t)
            print()
```
